```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
data = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
data.head()
```

```python
data['label'] = dataset.target
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data['label'].value_counts()
```

1 ---> Benign

0 ---> Malignant

```python
data.groupby('label').mean()
```

```python
X = data.drop(columns = 'label',axis = 1)
Y = data['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

Building the neural network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# creating the layers

model = keras.Sequential([
                            keras.layers.Flatten(input_shape = (30,)),
                            keras.layers.Dense(20, activation = 'relu'),
                            keras.layers.Dense(2, activation ='sigmoid')
])
```

```python
model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=15)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc = 'upper right')
```

```python
# accuracy on test data
loss,accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_pred = model.predict(X_test_std)
```

```python
print(Y_pred,Y_pred.shape)
```

Y_pred gives the prediction probability of each class for eac data point. We need to convert this to labels 0 or 1


```python
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

```python
# building a predictive system for a single input

input_data =(13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)
input_data = np.asarray(input_data)
input_data = input_data.reshape(1,-1)
input_data = scaler.transform(input_data)

prediction = model.predict(input_data)
prediction_label = [np.argmax(prediction)]
print(prediction)
print(prediction_label)
print('The tumor is Malignant' if prediction_label[0] == 0 else 'The tumor is Benign')
```
